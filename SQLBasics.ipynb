Q1. Create a New Database and Table for Employees

Task: Create a new database named company_db and Create a table named employees with the following
columns:
(Refer question paper for details)

```python
CREATE DATABASE company_db;

USE company_db;

CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    department VARCHAR(50),
    salary INT,
    hire_date DATE
);
```

Q3. Display All Employee Records Sorted by Salary (Lowest to Highest)

Hint: Use the ORDER BY
clause on the salary column.

```python
SELECT *
FROM employees
ORDER BY salary ASC;
```

Q4. Show Employees Sorted by Department (A–Z) and Salary (High → Low)

```python
SELECT *
FROM employees
ORDER BY department ASC, salary DESC;
```

Q5. List All Employees in the IT Department, Ordered by Hire Date (Newest First)

```python
SELECT *
FROM employees
WHERE department = 'IT'
ORDER BY hire_date DESC;
```

Q9. What is the Difference Between a Primary Key and a Foreign Key?

Primary Key

1. Uniquely identifies each record in a table
2. Cannot contain NULL values
3. Each table can have only one primary key


Foreign Key

1. Used to link one table to another
2. Refers to the primary key of another table
3. Can contain duplicate and NULL values


Q10. What Are Constraints in SQL and Why Are They Used?

Constraints are rules applied to table columns to ensure data accuracy and integrity.

Common SQL constraints:

* PRIMARY KEY – Unique identification of records
* NOT NULL – Prevents NULL values
* UNIQUE – Ensures all values are different
* CHECK – Limits the values allowed
* DEFAULT – Assigns a default value
* FOREIGN KEY – Maintains relationship between tables


Following are the reasons why they are used:

1. To maintain data consistency
2. To prevent invalid data entry
3. To enforce business rules
